```python
import collections
from bills.models.train import BillDataSet
from bills import consts

from torch.utils.data import DataLoader

dset = BillDataSet(consts.PROCESSED_DATA_DIR, os.path.join(consts.PROCESSED_DATA_DIR, "train.txt"))
dataloader = DataLoader(dset, batch_size=4, shuffle=True, num_workers=4)

classes = collections.Counter()
for batch in dataloader:
    classes.update(batch['status'])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
keys = classes.keys()
plt.barh(range(len(classes)), [classes[k] for k in keys], tick_label=keys,
         align='center')
```
